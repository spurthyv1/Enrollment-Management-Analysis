```python
NA_names = ['sex','avg_income','distance']
dfSV1[NA_names]=dfSV[NA_names].fillna(dfSV[NA_names].mean())
dfSV1['ETHNICITY'] = dfSV['ETHNICITY'].fillna('C')
dfSV1['TERRITORY'] = dfSV['TERRITORY'].fillna('2')
dfSV1.describe(include='all')
```

```python
dummy_data = pd.get_dummies(dfSV1[['ETHNICITY','TERRITORY']],drop_first=True)
```

```python
num_names=['SELF_INIT_CNTCTS','TRAVEL_INIT_CNTCTS',
           'SOLICITED_CNTCTS','REFERRAL_CNTCTS','CAMPUS_VISIT','sex',
           'mailq','premiere','interest','stucar','init_span','int1rat','int2rat',
           'hscrat','avg_income','distance']
x = pd.concat([dfSV1[num_names],dummy_data],axis=1)
x.describe(include='all')
```

```python
#create dependent variable
y = dfSV['Enroll']
```

```python
from sklearn.model_selection import train_test_split
xSV_train,xSV_val,ySV_train,ySV_val = train_test_split(x,y,
                                               test_size=0.3,
                                               random_state=0)
```

```python
import statsmodels.api as sm
log_regSV = sm.Logit(ySV_train,xSV_train).fit()
print(log_regSV.summary())
```

```python
from sklearn.metrics import confusion_matrix
prediction_prob = log_regSV.predict(xSV_val)   #predict based on val dataset
                                           # convert to binary numbers for confusion matrix

prediction = list(map(round,prediction_prob))  #default 0.5 threshold
confusion_matrix(ySV_val,prediction)             #compare actual behaviour(dep var-y) and model prediction
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot as plt

lr_auc = roc_auc_score(ySV_val,prediction_prob)
print('Logistic: ROC AUC=%.3f' % (lr_auc))
```

```python
# Build a decision tree model
from sklearn import tree
from sklearn.tree import export_text
dtreeSV = tree.DecisionTreeClassifier(max_depth=4,min_samples_split=30)
dtreeSV = dtreeSV.fit(xSV_train, ySV_train)
```

```python
# Display decision tree results
r = export_text(dtreeSV, feature_names=list(xSV_train.columns.values))
print(r)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=[25,20])
tree.plot_tree(dtreeSV,
               feature_names=list(xSV_train.columns.values),
               class_names=True,
               filled=True)
plt.show()
```

```python
#Evaluation model performance using the validation dataset
from sklearn.metrics import confusion_matrix
#Criteria 1: the confusion matrix
prediction =dtreeSV.predict(xSV_val)
confusion_matrix(ySV_val,prediction)
dtreeSV.score(xSV_val,ySV_val)
```

```python
#Criteria 2: the ROC curve and area under the curve
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
prediction_prob = dtreeSV.predict_proba(xSV_val)
```

```python
# calculate AUC scores
# only keeps the probabilities of one class, since tree models can be used for
tree_auc = roc_auc_score(ySV_val, prediction_prob[:,1])
print('Decision Tree: ROC AUC=%.3f' % (tree_auc))
```

```python
# calculate roc curves
tree_fpr, tree_tpr, _ = roc_curve(ySV_val, prediction_prob[:,1])
# plot the roc curve for the model
plt.plot(tree_fpr, tree_tpr, marker='.')
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# show the plot
# multi label classifications
plt.show()
```
